```python
from AMAT.planet import Planet
from AMAT.vehicle import Vehicle
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# Create a planet object
planet2=Planet("EARTH")
planet2.h_skip = 120000.0

# Load an nominal atmospheric profile with height, temp, pressure, density data
planet2.loadAtmosphereModel('../atmdata/earth-gram-avg.dat', 0 , 1 ,2, 3)
```

```python
# Set up a vehicle for drag modulation aerocapture and propagate the aerocapture trajectory
vehicle4=Vehicle('SmallSat1', 37, 20, 0.0, 1.767, 0.0, 0.35, planet2)
vehicle4.setInitialState(120.0,-6.22,23.16,10.8191,89.9982,-5.00,0.0,0.0)
vehicle4.setSolverParams(1E-6)
vehicle4.setDragModulationVehicleParams(20, 7.5)

# Set planet.h_low to 10 km, if vehicle dips below this level
# trajctory is terminated.
planet2.h_low=10.0E3

# Set target orbit = 2000 km x 2000 km, tolerance = 20 km
# target apo intentionally set to slightly higher value to account for bias in guidance
vehicle4.setTargetOrbitParams(200.0, 2000.0, 20.0)

# Set entry phase parameters
# v_switch_kms = 2.0, lowAlt_km = 20.0,
# numPoints_lowAlt = 101, hdot_threshold = -200.0 m/s.
# These are somewhat arbitary based on experience.
vehicle4.setDragEntryPhaseParams(2.0, 20.0, 101, -200.0)

# Set beta_1 and beta_ratio
vehicle4.setDragModulationVehicleParams(20, 7.5)

# Set vehicle initial state
vehicle4.setInitialState(120.0,-6.22,23.16,10.8191,89.9982,-4.40,0.0,0.0)
# Propogate a single vehicle trajectory
vehicle4.propogateGuidedEntryD2(1.0,1.0,0.1,2400.0)
```

```python
vehicle5 = Vehicle('PVenus', 91, 190.0, 0.0, np.pi*0.76**2.0*0.25, 0.0, 0.19, planet2)
vehicle5.setInitialState(120,0.0,0.0,11.0,0.0,-8.0,0.0,0.0)
vehicle5.setSolverParams(1E-6)
vehicle5.propogateEntry (2.0*60, 0.1, 0.0)
```

```python
vehicle6 = Vehicle('PVenus', 91, 190.0, 0.0, np.pi*0.76**2.0*0.25, 0.0, 0.19, planet2)
vehicle6.setInitialState(120,0.0,0.0,11.0,0.0,-15.0,0.0,0.0)
vehicle6.setSolverParams(1E-6)
vehicle6.propogateEntry (1.5*60, 0.1, 0.0)
```

```python
heatload_4 = integrate.cumtrapz(vehicle4.q_stag_total_full, vehicle4.t_min_full*60, initial=0)/1e3
heatload_5 = integrate.cumtrapz(vehicle5.q_stag_total, vehicle5.t_minc*60, initial=0)/1e3
heatload_6 = integrate.cumtrapz(vehicle6.q_stag_total, vehicle6.t_minc*60, initial=0)/1e3
```

```python
plt.figure(figsize=(11.25, 6.25))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.subplot(2, 2, 1)
plt.plot(vehicle5.t_minc, vehicle5.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -8$')
plt.plot(vehicle6.t_minc, vehicle6.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -15$')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 2)
plt.plot(vehicle5.stag_pres_atm, vehicle5.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -8$')
plt.plot(vehicle6.stag_pres_atm, vehicle6.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -15$')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.subplot(2, 2, 3)
plt.plot(vehicle4.t_min_full, vehicle4.q_stag_total_full, 'g-', linewidth=2.0)
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)

plt.subplot(2, 2, 4)
plt.plot(vehicle4.stag_pres_atm_full, vehicle4.q_stag_total_full, 'g-', linewidth=2.0)
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)


plt.savefig('../thermal-protection-systems-plots/earth.png', dpi= 300, bbox_inches='tight')
plt.show()
```
