```python
from AMAT.planet import Planet
from AMAT.vehicle import Vehicle
import numpy as np
import matplotlib.pyplot as plt
```

```python
planet8=Planet("NEPTUNE")
planet8.h_skip = 1000e3
planet8.loadAtmosphereModel('../atmdata/neptune-gram-avg.dat', 0 , 7 ,6, 5 , heightInKmFlag=True)
planet8.h_low = 120e3
planet8.h_trap= 0e3
```

```python
vehicle19A=Vehicle('Titania', 3200.0, 146 , 0.24, np.pi*4.5**2.0, 0.0, 1.125, planet8)
vehicle19A.setInitialState(1000.0,-15.22,75.55,30,8.687, -11.8822 ,0.0,0.0)
vehicle19A.setSolverParams(1E-6)
vehicle19A.propogateEntry (2400.0,0.1,180.0)
```

```python
vehicle19B=Vehicle('Titania', 3200.0, 146 , 0.24, np.pi*4.5**2.0, 0.0, 1.125, planet8)
vehicle19B.setInitialState(1000.0,-15.22,75.55,30,8.687, -12.6975 ,0.0,0.0)
vehicle19B.setSolverParams(1E-6)
vehicle19B.propogateEntry (2400.0,0.1,0.0)
```

```python
vehicle20=Vehicle('UOP', 268, 200, 0.0, np.pi*1.26**2.0*0.25, 0.0, 0.40, planet8)
vehicle20.setInitialState(1000,0.0,0.0,25,0.0,-30,0.0,0.0)
vehicle20.setSolverParams(1E-6)
vehicle20.propogateEntry(2*60, 0.1, 0.0)
```

```python
vehicle21=Vehicle('UOP', 268, 200, 0.0, np.pi*1.26**2.0*0.25, 0.0, 0.40, planet8)
vehicle21.setInitialState(1000,0.0,0.0,25,0.0,-40,0.0,0.0)
vehicle21.setSolverParams(1E-6)
vehicle21.propogateEntry(2*60, 0.1, 0.0)
```

```python
from scipy import integrate
```

```python
heatload_19A = integrate.cumtrapz(vehicle19A.q_stag_total, vehicle19A.t_minc*60, initial=0)/1e3
heatload_19B = integrate.cumtrapz(vehicle19B.q_stag_total, vehicle19B.t_minc*60, initial=0)/1e3

heatload_20 = integrate.cumtrapz(vehicle20.q_stag_total, vehicle20.t_minc*60, initial=0)/1e3
heatload_21 = integrate.cumtrapz(vehicle21.q_stag_total, vehicle21.t_minc*60, initial=0)/1e3
```

```python
plt.figure(figsize=(11.25, 6.25))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.subplot(2, 2, 1)
plt.plot(vehicle20.t_minc, vehicle20.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -30$')
plt.plot(vehicle21.t_minc, vehicle21.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -40$')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 2)
plt.plot(vehicle20.stag_pres_atm, vehicle20.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -30$')
plt.plot(vehicle21.stag_pres_atm, vehicle21.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -40$')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.subplot(2, 2, 3)
plt.plot(vehicle19A.t_minc, vehicle19A.q_stag_total, 'b-', linewidth=2.0, label='Overshoot')
plt.plot(vehicle19B.t_minc, vehicle19B.q_stag_total, 'r-', linewidth=2.0, label='Undershoot')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 4)
plt.plot(vehicle19A.stag_pres_atm, vehicle19A.q_stag_total, 'b-', linewidth=2.0, label='Overshoot')
plt.plot(vehicle19B.stag_pres_atm, vehicle19B.q_stag_total, 'r-', linewidth=2.0, label='Undershoot')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='lower right', fontsize=12)

plt.savefig('../thermal-protection-systems-plots/neptune.png', dpi= 300, bbox_inches='tight')
plt.show()
```
