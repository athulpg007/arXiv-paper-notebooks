```python
from AMAT.planet import Planet
from AMAT.vehicle import Vehicle
import numpy as np
import matplotlib.pyplot as plt
```

```python
planet7=Planet("URANUS")
planet7.h_skip = 1000e3
planet7.loadAtmosphereModel('../atmdata/uranus-gram-avg.dat', 0 , 1 ,2, 3, heightInKmFlag=True)
planet7.h_low = 120e3
planet7.h_trap= 0e3
```

```python
vehicle16A=Vehicle('Titania', 3200.0, 146 , 0.24, np.pi*4.5**2.0, 0.0, 1.125, planet7)
vehicle16A.setInitialState(1000.0,-15.22,75.55,29.2877,88.687,-11.0088 ,0.0,0.0)
vehicle16A.setSolverParams(1E-6)
vehicle16A.propogateEntry (2400.0,0.1,180.0)
```

```python
vehicle16B=Vehicle('Titania', 3200.0, 146 , 0.24, np.pi*4.5**2.0, 0.0, 1.125, planet7)
vehicle16B.setInitialState(1000.0,-15.22,75.55,29.2877,88.687,-12.0264 ,0.0,0.0)
vehicle16B.setSolverParams(1E-6)
vehicle16B.propogateEntry (2400.0,0.1,0.0)
```

```python
vehicle17=Vehicle('UOP', 268, 200, 0.0, np.pi*1.26**2.0*0.25, 0.0, 0.40, planet7)
vehicle17.setInitialState(1000,0.0,0.0,23,0.0,-30,0.0,0.0)
vehicle17.setSolverParams(1E-6)
vehicle17.propogateEntry(2*60, 0.1, 0.0)
```

```python
vehicle18=Vehicle('UOP', 268, 200, 0.0, np.pi*1.26**2.0*0.25, 0.0, 0.40, planet7)
vehicle18.setInitialState(1000,0.0,0.0,23,0.0,-50,0.0,0.0)
vehicle18.setSolverParams(1E-6)
vehicle18.propogateEntry(2*60, 0.1, 0.0)
```

```python
from scipy import integrate
```

```python
heatload_16A = integrate.cumtrapz(vehicle16A.q_stag_total, vehicle16A.t_minc*60, initial=0)/1e3
heatload_16B = integrate.cumtrapz(vehicle16B.q_stag_total, vehicle16B.t_minc*60, initial=0)/1e3

heatload_17 = integrate.cumtrapz(vehicle17.q_stag_total, vehicle17.t_minc*60, initial=0)/1e3
heatload_18 = integrate.cumtrapz(vehicle18.q_stag_total, vehicle18.t_minc*60, initial=0)/1e3
```

```python
plt.figure(figsize=(11.25, 6.25))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.subplot(2, 2, 1)
plt.plot(vehicle17.t_minc, vehicle17.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -30$')
plt.plot(vehicle18.t_minc, vehicle18.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -50$')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 2)
plt.plot(vehicle17.stag_pres_atm, vehicle17.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -30$')
plt.plot(vehicle18.stag_pres_atm, vehicle18.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -50$')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.subplot(2, 2, 3)
plt.plot(vehicle16A.t_minc, vehicle16A.q_stag_total, 'b-', linewidth=2.0, label='Overshoot')
plt.plot(vehicle16B.t_minc, vehicle16B.q_stag_total, 'r-', linewidth=2.0, label='Undershoot')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 4)
plt.plot(vehicle16A.stag_pres_atm, vehicle16A.q_stag_total, 'b-', linewidth=2.0, label='Overshoot')
plt.plot(vehicle16B.stag_pres_atm, vehicle16B.q_stag_total, 'r-', linewidth=2.0, label='Undershoot')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='lower right', fontsize=12)

plt.savefig('../thermal-protection-systems-plots/uranus.png', dpi= 300, bbox_inches='tight')
plt.show()
```

```python
print(f'{round(heatload_16A[-1], 2)} kJ/cm2')
print(f'{round(heatload_16B[-1], 2)} kJ/cm2')
print('---')
print(f'{round(heatload_17[-1], 2)} kJ/cm2')
print(f'{round(heatload_18[-1], 2)} kJ/cm2')
```
