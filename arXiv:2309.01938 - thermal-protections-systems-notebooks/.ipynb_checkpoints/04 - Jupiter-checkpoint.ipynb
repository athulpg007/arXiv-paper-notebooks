```python
from AMAT.planet import Planet
from AMAT.vehicle import Vehicle
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# Set up the planet and atmosphere model.
planet4=Planet("JUPITER")
planet4.h_skip = 1000E3
planet4.h_trap =  0.0E3
planet4.loadAtmosphereModel('../atmdata/jupiter-galileo-asi.dat', 0 , 1 , 2, 3, heightInKmFlag=True)
```

```python
vehicle10=Vehicle('Galileo', 340, 267, 0.0, np.pi*1.26**2.0*0.25, 0.0, 0.22, planet4)
vehicle10.setInitialState(450,0.0,0.0,47.4,0.0,-8.00,0.0,0.0)
vehicle10.setSolverParams(1E-6)
vehicle10.propogateEntry (2*60, 0.1, 0.0)
```

```python
heatload_10 = integrate.cumtrapz(0.4*vehicle10.q_stag_total, vehicle10.t_minc*60, initial=0)/1e3
```

```python
print(f'{round(heatload_10[-1], 2)} kJ/cm2')
```

```python
plt.figure(figsize=(12.25, 6.25))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.subplot(2, 2, 1)
plt.plot(vehicle10.t_minc, 0.4*vehicle10.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -8$')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 2)
plt.plot(vehicle10.stag_pres_atm, 0.4*vehicle10.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -8$')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.savefig('../thermal-protection-systems-plots/jupiter.png', dpi= 300, bbox_inches='tight')
plt.show()
```
