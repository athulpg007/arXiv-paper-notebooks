```python
from AMAT.planet import Planet
from AMAT.vehicle import Vehicle
import numpy as np
import matplotlib.pyplot as plt
```

```python
planet6 = Planet('TITAN')
planet6.loadAtmosphereModel('../atmdata/titan-gram-avg.dat', 0 , 1 ,2, 3)
planet6.h_skip = 1000.0E3
```

```python
vehicle13=Vehicle('TitanOrbiter', 5700, 30, 0.0, 113.0, 0.0, 1.00, planet6)
vehicle13.setInitialState(1000, -23.67, 14.89, 7.3384, 90.1226, -5.00, 0.0,0.0)
vehicle13.setSolverParams(1E-5)
vehicle13.setDragModulationVehicleParams(30, 4.14)
```

```python
# Set planet.h_low to 10 km, if vehicle dips below this level
# trajctory is terminated.
planet6.h_low=50.0E3

# Set target orbit = 2000 km x 2000 km, tolerance = 20 km
# target apo intentionally set to slightly higher value to account for bias in guidance
vehicle13.setTargetOrbitParams(1700, 1700, 20.0)

# Set entry phase parameters
# v_switch_kms = 2.0, lowAlt_km = 20.0,
# numPoints_lowAlt = 101, hdot_threshold = -200.0 m/s.
# These are somewhat arbitary based on experience.
vehicle13.setDragEntryPhaseParams(2.0, 50.0, 101, -200.0)

# Set beta_1 and beta_ratio
vehicle13.setDragModulationVehicleParams(30, 4.14)

# Set vehicle initial state
vehicle13.setInitialState(1000, -23.67, 14.89, 7.3384, 90.1226, -35.0, 0.0,0.0)
# Propogate a single vehicle trajectory
vehicle13.propogateGuidedEntryD2(0.5,1.0,1.0,6000.0)
```

```python
vehicle14=Vehicle('Huygens', 320, 35, 0.0, np.pi*2.7**2.0*0.25, 0.0, 1.25, planet6)
vehicle14.setInitialState(1000,0.0,0.0,7.0,0.0,-45,0.0,0.0)
vehicle14.setSolverParams(1E-6)
vehicle14.propogateEntry (5*60, 1.0, 0.0)
```

```python
vehicle15=Vehicle('Huygens', 320, 35, 0.0, np.pi*2.7**2.0*0.25, 0.0, 1.25, planet6)
vehicle15.setInitialState(1000,0.0,0.0,7.0,0.0,-65,0.0,0.0)
vehicle15.setSolverParams(1E-6)
vehicle15.propogateEntry (5*60, 1.0, 0.0)
```

```python
from scipy import integrate
```

```python
heatload_13 = integrate.cumtrapz(vehicle13.q_stag_total_full, vehicle13.t_min_full*60, initial=0)/1e3
heatload_14 = integrate.cumtrapz(vehicle14.q_stag_total, vehicle14.t_minc*60, initial=0)/1e3
heatload_15 = integrate.cumtrapz(vehicle15.q_stag_total, vehicle15.t_minc*60, initial=0)/1e3
```

```python
plt.figure(figsize=(11.25, 6.25))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.subplot(2, 2, 1)
plt.plot(vehicle14.t_minc, vehicle14.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -45$')
plt.plot(vehicle15.t_minc, vehicle15.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -65$')
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper right', fontsize=12)

plt.subplot(2, 2, 2)
plt.plot(vehicle14.stag_pres_atm, vehicle14.q_stag_total, 'b-', linewidth=2.0, label=r'$\gamma = -45$')
plt.plot(vehicle15.stag_pres_atm, vehicle15.q_stag_total, 'r-', linewidth=2.0, label=r'$\gamma = -65$')
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.subplot(2, 2, 3)
plt.plot(vehicle13.t_min_full, vehicle13.q_stag_total_full, 'g-', linewidth=2.0)
plt.xlabel('Time, min',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)

plt.subplot(2, 2, 4)
plt.plot(vehicle13.stag_pres_atm_full, vehicle13.q_stag_total_full, 'g-', linewidth=2.0)
plt.xlabel('Stagnation pressure, atm',fontsize=12)
plt.ylabel('Heat rate, '+r'$W/cm^2$',fontsize=12)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=12)
ax.tick_params(axis='y',labelsize=12)
plt.grid(linestyle='dotted', linewidth=0.5)


plt.savefig('../thermal-protection-systems-plots/titan.png', dpi= 300, bbox_inches='tight')
plt.show()
```
