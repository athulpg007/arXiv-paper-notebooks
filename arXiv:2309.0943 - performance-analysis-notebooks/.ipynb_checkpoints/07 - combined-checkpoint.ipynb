```python
import numpy as np
from AMAT.approach import Approach
from AMAT.orbiter import PropulsiveOrbiter
import matplotlib.pyplot as plt
```

```python
def mf_m0(DV, Isp=320):
    return  round(max(0.01, 1 - 1.10*(1 - np.exp(-DV/(Isp*9.80665)))), 2)
```

```python
approach1 = Approach(
    "VENUS", 
    v_inf_vec_icrf_kms=np.array([-3.26907095,  0.67649492, -1.06977471]),
    rp=(6051.8+400)*1e3, 
    psi=np.pi
)

orbiter1 = PropulsiveOrbiter(approach=approach1, apoapsis_alt_km=400)
```

```python
approach2 = Approach(
    "EARTH", 
    v_inf_vec_icrf_kms=np.array([-3.26907095,  0.67649492, -1.06977471]),
    rp=(6378+400)*1e3, 
    psi=np.pi
)

orbiter2 = PropulsiveOrbiter(approach=approach2, apoapsis_alt_km=400)
```

```python
approach3 = Approach(
    "MARS", 
    v_inf_vec_icrf_kms=np.array([ 2.23930482,  1.20086474, -0.73683367]),
    rp=(3389.5+400)*1e3, 
    psi=np.pi
)

orbiter3 = PropulsiveOrbiter(approach=approach3, apoapsis_alt_km=400)
```

```python
approach4 = Approach(
    "TITAN", 
    v_inf_vec_icrf_kms=np.array([-0.910, 5.081, 4.710]),
    rp=(2575 + 1700)*1e3, 
    psi=np.pi
)

orbiter4 = PropulsiveOrbiter(approach=approach4, apoapsis_alt_km=1700)
```

```python
approach5A = Approach(
    "URANUS", 
    v_inf_vec_icrf_kms=np.array([0.1, 10.01192666,  0.1]),
    rp=(25559 + 4000)*1e3, 
    psi=np.pi
)

approach5B = Approach(
    "URANUS", 
    v_inf_vec_icrf_kms=np.array([0.1, 20.01192666,  0.1]),
    rp=(25559 + 4000)*1e3, 
    psi=np.pi
)

orbiter5A = PropulsiveOrbiter(approach=approach5A, apoapsis_alt_km=1000E3)
orbiter5B = PropulsiveOrbiter(approach=approach5B, apoapsis_alt_km=1000E3)
```

```python
approach6A = Approach(
    "NEPTUNE", 
    v_inf_vec_icrf_kms=np.array([0.1, 10.01192666,  0.1]),
    rp=(25559 + 4000)*1e3, 
    psi=np.pi
)

approach6B = Approach(
    "NEPTUNE", 
    v_inf_vec_icrf_kms=np.array([0.1, 20.01192666,  0.1]),
    rp=(25559 + 4000)*1e3, 
    psi=np.pi
)

orbiter6A = PropulsiveOrbiter(approach=approach6A, apoapsis_alt_km=500E3)
orbiter6B = PropulsiveOrbiter(approach=approach6B, apoapsis_alt_km=500E3)
```

```python
y1 = [
    mf_m0(orbiter1.DV_OI_mag), 
    mf_m0(orbiter2.DV_OI_mag),
    mf_m0(orbiter3.DV_OI_mag),
    mf_m0(orbiter4.DV_OI_mag),
    mf_m0(orbiter5A.DV_OI_mag),
    mf_m0(orbiter5B.DV_OI_mag),
    mf_m0(orbiter6A.DV_OI_mag),
    mf_m0(orbiter6B.DV_OI_mag)
]

y2 = [0.50, 0.50, 0.55, 0.50, 0.50, 0.50, 0.50, 0.50]
```

```python
labels = (np.array(y2) - np.array(y1))*100/np.array(y1)
labels = [f'+ {round(label)}%' if label < 1000 else np.infty for label in labels]
```

```python
plt.figure(figsize=(12.25, 5.5))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

x = np.arange(8)
# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width=0.4, color='red')
plt.bar(x+0.2, y2, width=0.4, color='blue', hatch='/')


plt.xticks(x, [
    'Venus', 
    'Earth', 
    'Mars', 
    'Titan', 
    'Uranus\n(slow arrival)', 
    'Uranus\n(fast arrival)',
    'Neptune\n(slow arrival)', 
    'Neptune\n(fast arrival)',
])

for (xx, y, label) in zip(x, y2, labels):
    plt.text(xx + 0.20, y, label, ha='center', va='bottom', fontsize=14, weight='bold')
    
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=14)
ax.tick_params(axis='y',labelsize=14)
plt.grid(linestyle='dotted', linewidth=0.5)
plt.ylabel('Mass fraction delivered to orbit', fontsize=16)
plt.ylim([0, 0.70])

plt.legend(["Propulsive insertion", "Aerocapture"], loc='upper right', fontsize=14)
plt.savefig('../performance-analysis-plots/combined.png', dpi= 300, bbox_inches='tight')
plt.show()
```
