```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
data_max = np.loadtxt('../data/neptune-fminmax/FMINMAX+10-small.txt', usecols=[0, 3])
data_avg = np.loadtxt('../data/neptune-fminmax/FMINMAX+00-small.txt', usecols=[0, 2])
data_min = np.loadtxt('../data/neptune-fminmax/FMINMAX-10-small.txt', usecols=[0, 1])
```

```python
height = data_min[:, 0]

dens_low = data_min[:,1]
dens_avg = data_avg[:,1]
dens_high = data_max[:,1]
```

```python
dens_low_ratio = dens_low/dens_avg
dens_high_ratio = dens_high/dens_avg
```

```python
fig = plt.figure()
fig.set_size_inches([12.25, 5.25])

plt.subplot(1, 2, 1)
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.plot(dens_low, height, 'b-', linewidth=1.5, label="FMINMAX = -1")
plt.plot(dens_avg, height, 'g-', linewidth=1.5, label="FMINMAX = +0")
plt.plot(dens_high, height, 'r-', linewidth=1.5, label="FMINMAX = +1")
plt.xlabel("Density, kg/m3",fontsize=14)
plt.ylabel("Altitude, km",fontsize=14)
plt.xscale('log')
plt.yticks(fontsize=14)
plt.xticks(np.logspace(-10, 0, 11), fontsize=14)
plt.grid('on',linestyle='-', linewidth=0.2)

ax=plt.gca()
ax.xaxis.set_tick_params(direction='in', which='both')
ax.yaxis.set_tick_params(direction='in', which='both')
ax.xaxis.set_tick_params(width=2, length=4)
ax.yaxis.set_tick_params(width=2, length=4)
ax.xaxis.set_tick_params(width=1, length=4, which='minor')
ax.yaxis.set_tick_params(width=1, length=4, which='minor')
ax.xaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.xaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.0)
ax.yaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.yaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.0)
ax.add_patch(Rectangle((1E-10, 200), 1, 200, alpha=0.5, facecolor='xkcd:sky blue'))

for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)

plt.legend(loc='upper right', fontsize=12, framealpha=0.8)

plt.subplot(1, 2, 2)
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.plot(dens_low_ratio, height, 'b-', linewidth=2.0, label="FMINMAX = -1")
plt.plot(dens_high_ratio, height, 'r-', linewidth=2.0, label="FMINMAX = +1")
plt.xlabel("Multiplication factor",fontsize=14)
plt.ylabel("Altitude, km",fontsize=14)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.xscale('log')
plt.grid('on',linestyle='-', linewidth=0.2)

ax=plt.gca()
ax.xaxis.set_tick_params(direction='in', which='both')
ax.yaxis.set_tick_params(direction='in', which='both')
ax.xaxis.set_tick_params(width=2, length=4)
ax.yaxis.set_tick_params(width=2, length=4)
ax.xaxis.set_tick_params(width=1, length=4, which='minor')
ax.yaxis.set_tick_params(width=1, length=4, which='minor')
ax.xaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.xaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)
ax.yaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.yaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)
ax.add_patch(Rectangle((-1500, 200), 2000, 200, alpha=0.5, facecolor='xkcd:sky blue'))

for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)

plt.legend(loc='upper center', fontsize=12, framealpha=0.8)
plt.savefig('../atmospheric-uncertainties-plots/neptune-fminmax.png', dpi=300, bbox_inches='tight')
plt.show()
```
